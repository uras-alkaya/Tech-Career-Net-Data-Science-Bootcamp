#**1. Numpy Part**

**- Matrix Operations with Numpy**

```python
import numpy as np

# Create a 5x5 random integer matrix
matrix = np.random.randint(0, 101, size=(5, 5))
print("Matrix:\n", matrix, "\n")

# Mean, Standard Deviation, Variance
mean_val = np.mean(matrix)
std_val = np.std(matrix)
var_val = np.var(matrix)
print("Mean:", mean_val)
print("Standard Deviation:", std_val)
print("Variance:", var_val, "\n")

# Max and Min values
max_val = np.max(matrix)
min_val = np.min(matrix)
print("Max value:", max_val)
print("Min value:", min_val, "\n")

# Sum of diagonal elements
diag_sum = np.trace(matrix)
print("Sum of diagonal elements:", diag_sum)
```

**- Data Simulation with Numpy**

```python
# Simulate exam scores of 1000 students
scores = np.random.normal(loc=70, scale=15, size=1000)
scores = np.clip(scores, 0, 100)

# Mean, Median, Std
mean_score = np.mean(scores)
median_score = np.median(scores)
std_score = np.std(scores)
print("Average Score:", mean_score)
print("Median Score:", median_score)
print("Standard Deviation:", std_score, "\n")

# Number of students scoring below 50
below_50 = np.sum(scores < 50)
print("Number of students scoring below 50:", below_50)
```

#**2. Pandas Part**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create the dataset as a DataFrame
data = {
    "Student": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Age": [20, 21, 19, 22, 20],
    "Department": ["Computer Science", "Physics", "Chemistry", "Computer Science", "Physics"],
    "Math": [70, 60, 80, 90, 55],
    "Physics": [65, 75, 70, 85, 60],
    "Chemistry": [80, 85, 65, 95, 70]
}
```

```python
df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

```python
# 2. Find the average score for each subject
avg_scores = df[["Math", "Physics", "Chemistry"]].mean()
print("Subject Averages:\n", avg_scores)
```

```python
# 3. Find the student with the highest math score
highest_math_student = df.loc[df["Math"].idxmax(), "Student"]
print("Student with the highest Math score:", highest_math_student)
```

```python
# 4. Add a new column calculating each student’s average score
df["Average"] = df[["Math", "Physics", "Chemistry"]].mean(axis=1)
print("DataFrame with Average column:\n", df)
```

```python
# 5. Group by department and calculate the average performance
dept_averages = df.groupby("Department")[["Math", "Physics", "Chemistry", "Average"]].mean()
print("Department Averages:\n", dept_averages)
```

```python
# 6. Filter students with an average score above 70
above_70 = df[df["Average"] > 70]
print("Students with Average above 70:\n", above_70)
```

#**3. Task**

```python
# Histogram for each subject
plt.figure(figsize=(12, 4))
for i, subject in enumerate(["Math", "Physics", "Chemistry"], 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[subject], bins=5, kde=True, color="turquoise")
    plt.title(f"{subject} Score Distribution")
plt.tight_layout()
plt.show()
```

```python
# Barplot for department averages
plt.figure(figsize=(8, 5))
dept_averages[["Math", "Physics", "Chemistry", "Average"]].plot(
    kind="bar", figsize=(10, 6), colormap="coolwarm"
)
plt.title("Average Scores by Department")
plt.ylabel("Average Score")
plt.xlabel("Department")
plt.xticks(rotation=45)
plt.legend(title="Subjects")
plt.tight_layout()
plt.show()
```
